### Tech note
Prefer using `np.multiply`, `np.add`, `np.divide`, `np.subtract` instead of `*`,`+`,`/`,`-` for better memory handling.

Example: suppose you allocated a variable 

```
a = np.zeros(...)
```
So, instead of
```
a = b + c  # will be reallocated, GC needed to free
``` 
You can use: 
```
np.add(b,c,out = a) # puts result in `a`
```

````python
# batch generator
def get_batches(dataset, batch_size):
    X, Y = dataset
    n_samples = X.shape[0]
        
    # Shuffle at the start of epoch
    indices = np.arange(n_samples)
    np.random.shuffle(indices)
    
    for start in range(0, n_samples, batch_size):
        end = min(start + batch_size, n_samples)
        
        batch_idx = indices[start:end]
    
        yield X[batch_idx], Y[batch_idx]
````

````python
def accuracy(predictions, y_batch):
    return np.mean(np.argmax(predictions, axis=1) == np.argmax(y_batch, axis=1))
````

````python
def train(net, optimizer_config, optimizer_state, optimizaer, n_epoch, batch_size, criterion):
    loss_history = []
    accurcy_history = []
    bar = tqdm(total=n_epoch)
    for i in range(n_epoch):
        bar.update()
        # Test loss calculation
        net.train()
        for x_batch, y_batch in get_batches((X_train_flatten, y_train_onehot), batch_size):

            net.zeroGradParameters()

            # Forward
            predictions = net.forward(x_batch)
            loss = criterion.forward(predictions, y_batch)

            # Backward
            dp = criterion.backward(predictions, y_batch)
            net.backward(x_batch, dp)


            # Update weights
            optimizaer(net.getParameters(), 
                         net.getGradParameters(), 
                         optimizer_config,
                         optimizer_state)   
        net.evaluate()
        test_predictions = net.forward(X_test_flatten)
        test_loss = criterion.forward(test_predictions, y_test_onehot)
        test_accuracy = accuracy(test_predictions, y_test_onehot)
        loss_history.append(test_loss)
        accurcy_history.append(test_accuracy)
    return loss_history, accurcy_history

def visualize(loss_history, accurcy_history, name):
    # Visualize
    display.clear_output(wait=True)
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

    f.suptitle(name + " activation function")
    ax1.set_title("Test loss")
    ax1.set_xlabel("#iteration")
    ax1.set_ylabel("loss")
    ax1.plot(loss_history, 'b')

    ax2.set_title("Test accuracy")
    ax2.set_xlabel("#iteration")
    ax2.set_ylabel("accuracy")
    ax2.plot(accurcy_history, 'b')
    plt.show()

    print('Current accuracy: %f' % accurcy_history[-1])
    print('Current loss: %f' % loss_history[-1])
````

````python
def plot_train_process(train_loss, val_loss, val_accuracy):
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    axes[0].set_title('Loss')
    axes[0].plot(train_loss, label='train')
    axes[0].plot(val_loss, label='validation')
    axes[0].legend()

    axes[1].set_title('Validation accuracy')
    axes[1].plot(val_accuracy)
````

````python
def get_batches(dataset, batch_size):
    X, Y = dataset
    n_samples = X.shape[0]
        
    # Shuffle at the start of epoch
    indices = torch.randperm(n_samples)
    
    for start in range(0, n_samples, batch_size):
        end = min(start + batch_size, n_samples)
        
        batch_idx = indices[start:end]
    
        yield X[batch_idx], Y[batch_idx]
````

````python
def accuracy(predictions, y_batch):
    return np.mean(np.argmax(predictions, axis=1) == np.argmax(y_batch, axis=1))
````

````python
def accuracy(predictions, y_batch):
    return np.mean(np.argmax(predictions, axis=1) == np.argmax(y_batch, axis=1))
````

````python
model = nn.Sequential(
    nn.Dropout(0.3),
    nn.Linear(pixels, 256),
    nn.LeakyReLU(),
    nn.BatchNorm1d(256, momentum=0),
    nn.Linear(256, 128),
    nn.LeakyReLU(),
    nn.BatchNorm1d(128, momentum=0),
    nn.Linear(128, 32),
    nn.BatchNorm1d(32, momentum=0),
    nn.Linear(32, 10),
    nn.LogSoftmax(dim=1)
)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device, torch.float32)
opt = torch.optim.Adam(model.parameters(), lr=3e-3, betas=(0.7, 0.9))

loss_func = nn.NLLLoss()
````

````python
n_epoch = 120
batch_size = 256

bar = tqdm(total=n_epoch)
accuracy_list = []
train_loss = []
for i in range(n_epoch):
    bar.update()
    # Test loss calculation
    model.train(True)
    for x_batch, y_batch in get_batches((X_train_flatten, y_train_onehot), batch_size):
        # Forward
        predictions = model(x_batch)
        #print(torch.argmax(y_batch, dim=1))
        loss = loss_func(predictions, torch.argmax(y_batch, dim=1))

        # Backward
        loss.backward()

        # Update weights
        opt.step()
        opt.zero_grad()
    train_loss.append(loss)

    model.train(False)
    val_predictions = model(X_val_flatten)
    accuracy_list.append(accuracy(val_predictions.cpu().detach().numpy(), y_val_onehot.cpu().detach().numpy()))
test_predictions = model(X_test_flatten)
test_loss = loss_func(test_predictions, torch.argmax(y_test_onehot, dim=1))
test_accuracy = accuracy(test_predictions.cpu().detach().numpy(), y_test_onehot.cpu().detach().numpy())
````
